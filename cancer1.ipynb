# Step 1 : Importing the libraries

```python
import numpy as np
import pandas as pd
```

# Step 2 : Reading the dataset

```python
dataset = pd.read_csv("cancer_dataset (1).csv") # first we need upload the dataset
dataset.head() # will shoe first 5 rows
```

Step 3 : Seperating the features and target¶

```python
x = dataset.iloc[:,2:32]
y = dataset.iloc[:,1]
x.head()
```

```python
y.head()
```

# Step 4 : Splitting the dataset into training set and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

# Step 5 : Creating the Model

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

# Step 6 : Training Model 

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test) # predicting for all x_test
y_pred
```

# Step 7 : Testing the Model

```python
x_test
```

y_test

```python
list(zip(y_test,y_pred))
```

# Step 8 : Check for accuracy

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
print(acc*100)
```

# Step 9 : Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Assuming y_test and y_pred are already defined
cm = confusion_matrix(y_test, y_pred)
# Visualization with Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
